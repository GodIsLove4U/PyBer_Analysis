```python
%matplotlib inline

# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load in csv
pyber_ride_df = pd.read_csv("../PyBer_Analysis/ride_data.csv")
pyber_city_df = pd.read_csv("../PyBer_Analysis/city_data.csv")
```

```python
# Read city data
city_data = pyber_city_df
city_data.head()
```

```python
# Read ride data
ride_data = pyber_ride_df
ride_data.head()
```

```python
# Combine info from data set
combined_df = pd.merge(city_data, ride_data, how='outer', on='city')
combined_df.head()
```

```python
# Calculate totals and averages
# Fares, rides, drivers and city types
```

```python
city_type_count = "city" + "ride_id"
city_type_count
```

```python
# Fares
avg_fare = combined_df.groupby("city")["fare"].mean()
avg_fare = pd.DataFrame(avg_fare).reset_index()
avg_fare = avg_fare.rename(columns = {'fare' : 'average fare'})
avg_fare.head()
```

```python
combined_fareinfo_df = pd.merge(combined_df, avg_fare, how='outer', on='city')
combined_fareinfo_df
```

```python
# Rides
city_rides = combined_df.groupby("city")["ride_id"].count()
city_rides = pd.DataFrame(city_rides).reset_index()
city_rides = city_rides.rename(columns = {'ride_id' : 'total rides'})
city_rides.head()
```

```python
# Drivers
city_drivers = combined_df[["city" , "driver_count"]].drop_duplicates("city")
city_drivers = pd.DataFrame(city_drivers).reset_index()
city_drivers.head()
```

```python
# City types
city_type = combined_df[["city" , "type"]].drop_duplicates("city")
city_type = pd.DataFrame(city_type).reset_index()
city_type.head()
```

```python
rideshare_df = pd.merge(pd.merge(pd.merge(avg_fare, city_rides, on="city"), 
                            city_drivers, on="city"), city_type, on="city")
rideshare_df.head()
```

```python
rideshare_df.columns = ['City','Average Fare per Ride','Total Rides','X','Total Drivers','Y','Type']
rideshare_df
```

```python
# Fares by city type
avg_fare_type = combined_df.groupby("type")["fare"].mean()
avg_fare_type = pd.DataFrame(avg_fare_type).reset_index()
avg_fare_type = avg_fare_type.rename(columns = {'fare' : 'average fare type'})
avg_fare_type.head()
```

```python
# Fares by driver
avg_fare_driver = combined_df.groupby("ride_id")["fare"].mean()
avg_fare_driver = pd.DataFrame(avg_fare_driver).reset_index()
avg_fare_driver = avg_fare_driver.rename(columns = {'fare' : 'average driver fare'})
avg_fare_driver
```
